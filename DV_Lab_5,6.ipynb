The scatter plot matrix can be plotted using 3 tools: Pandas, Seaborn and Plotly.express. Scatter plot matrix is plotted using all the 3 tools below.

```python
# Scatter plot matrix using Pandas
colors = {"Setosa":'orange', "Virginica":'red', "Versicolor":'green'}
f = pd.plotting.scatter_matrix(iris_df,figsize=(15,15),grid=True, marker='^', color=iris_df['species'].map(colors))
```

```python
# Scatter plot matrix using plotly.express
f = px.scatter_matrix(iris_df)
f.show()
```

```python
g = px.scatter_matrix(iris_df,
    dimensions=iris_df.columns[0:4],
    color="species", symbol="species",
    labels=iris_df.columns) 
g.show()
```

```python
# Scatter plot matrix using Seaborn
sns.set_theme(style="ticks")
sns.pairplot(iris_df, hue='species')
```

```python
corr = iris_df.corr()
print(corr)
```

From the above scatter plot matrix and the correlation coeffients we can observe that:
*   Petal length and petal width are strongly positively correlated, with a correlation coefficient of 0.96.
*   Sepal length and petal length appear to be positively correlated, with a correlation coefficient of 0.87.
*   Sepal length and petal width also appear to be positively correlated, with a correlation coefficient of 0.82.
*   Sepal width does not appear to be strongly correlated with any of the other variables but is negatively correlated with all the variables.

**Which classes is linearly seperable?**

Of all the three classes i.e Setosa, Versicolor and Virginica, from the plotted scatterplot matrix we can observe that the ***Setosa*** class appears to be linealy separable from the remaining two classes i.e. Versicolor and virginica as it can be clearly seperated from the remaining two classes in all the plots
